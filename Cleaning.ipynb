```python
import pandas as pd
```

```python
# Load data
df = pd.read_csv('BenettonSuppliers.csv')
```

## Product types

```python
# Replace the values in the column 'APPAREL', 'ACCESSORIES' and 'SHOES' with True if the content is l 
# False if the entire content of the cell is empty


df['APPAREL'] = df['APPAREL'].replace('l', True)
df['APPAREL'] = df['APPAREL'].fillna(False)
df['ACCESSORIES'] = df['ACCESSORIES'].replace('l', True)
df['ACCESSORIES'] = df['ACCESSORIES'].fillna(False)
df['SHOES'] = df['SHOES'].replace('l', True)
df['SHOES'] = df['SHOES'].fillna(False)

df.to_csv('BenettonSuppliersCleaned1.0.csv', index=False)

# render the table in a browsable way
from IPython.display import display
display(df)

```

```python
# Same thing for the number of ; = 1
df[df['ADDRESS'].str.count(';') == 1]
```

```python
# I fixed the irregularities in libreoffice calc and I run again the notebook to load the new csv file.
# Split the 'ADDRESS' column into 'STREET', 'CITY' and 'ZIP' columns based on the ; separator, except 
# for the rows that have 1 or 3 ; in the column 'ADDRESS'

df['STREET'] = df['ADDRESS'].str.split(';').str[0]
df['CITY'] = df['ADDRESS'].str.split(';').str[1]
df['ZIP'] = df['ADDRESS'].str.split(';').str[2]
```

```python
# Split the column 'ADDRESS' into 'ADDRESS' and 'COUNTY', by extracting the words in brackets and putting them in a new column. Remove the brackets and the words in them from the 'ADDRESS' column.
df['COUNTY'] = df['ADDRESS'].str.extract(r'\((.*?)\)')
df['ADDRESS'] = df['ADDRESS'].str.replace(r'\(.*?\)','')
df.head()
```

```python
# Print column names
df.columns

df = df[['COMPANY NAME',  'VENDOR', 'COUNTRY', 'COUNTY', 'ZIP', 'CITY', 'STREET', 'ADDRESS',  'APPAREL',
         'ACCESSORIES', 'SHOES', 'PRODUCTION PHASE', 'EMPLOYEES RANGE', '% MEN',
         '% WOMEN', '% MIGRANT WORKERS', 'TRADE UNION',
         'NAME OF THE TRADE UNION', 'COLLECTIVE BARGAINING AGREEMENT',
         '% WORKFORCE COVERED BY THE CBA', 'WORKERS\' REPRESENTATIVES',
         'CERTIFICATIONS']]

df.head()
```

```python
# From the ZIP column, extract the terms in brackets and put them in a new column called 'COUNTY'
df['COUNTY'] = df['ZIP'].str.extract(r'\((.*?)\)')

df['ZIP'] = df['ZIP'].str.replace(r'\(.*?\)', '', regex=True)
df.head()

```

```python
# Remove the ADDRESS column
df = df.drop(['ADDRESS'], axis=1)
df.head()
```

```python
# Save the cleaned data in a new csv file
df.to_csv('BenettonSuppliersCleaned2.0.csv', index=False)
```

```python
# Represent the dataset using the display function and showing all of the columns.
from IPython.display import display
pd.options.display.max_columns = None
display(df)
```

## Production phase

```python
# How many values can there be in the PRODUCTION PHASE column if we split the string by ;? List the unique values

# Extract unique values in the 'PRODUCTION PHASE' column after splitting by ';'
unique_production_phases = set()
for phases in df['PRODUCTION PHASE'].dropna():
    for phase in phases.split(';'):
        unique_production_phases.add(phase.strip())

# Count of unique production phases and the list
count_unique_production_phases = len(unique_production_phases)
unique_production_phases, count_unique_production_phases
```

```python

# New columns in df: 
df = df.assign(MANUFACTURING=False, OTHER_PROCESSING=False, PRINTING=False, SPINNING=False, WEAVING=False, WET_PROCESS=False,)
phases = ['MANUFACTURING', 'OTHER_PROCESSING', 'PRINTING', 'SPINNING', 'WEAVING', 'WET_PROCESS']
for index, phases in df['PRODUCTION PHASE'].items():
    supplier_phases = phases.split(';')    
    if len(supplier_phases) > 0:
        for phase in supplier_phases:   
            phase = phase.strip().replace(" ", "_")         
            df[phase][index] = True
        

df.head()

```

```python

df = df[['COMPANY NAME',  'VENDOR', 'COUNTRY', 'COUNTY', 'ZIP', 'CITY', 'STREET',  'APPAREL',
         'ACCESSORIES', 'SHOES', 'MANUFACTURING', 'OTHER_PROCESSING', 'PRINTING', 'SPINNING', 'WEAVING', 'WET_PROCESS', 'EMPLOYEES RANGE', '% MEN',
         '% WOMEN', '% MIGRANT WORKERS', 'TRADE UNION',
         'NAME OF THE TRADE UNION', 'COLLECTIVE BARGAINING AGREEMENT',
         '% WORKFORCE COVERED BY THE CBA', 'WORKERS\' REPRESENTATIVES',
         'CERTIFICATIONS']]

df.to_csv('BenettonSuppliersCleaned3.0.csv', index=False)

df.head()
```

## Employee range

```python
# Initialize the new columns with NaN (Not a Number)
df['MIN_EMPLOYEES'] = pd.NA
df['MAX_EMPLOYEES'] = pd.NA

# Iterate over the DataFrame and update 'MIN_EMPLOYEES' and 'MAX_EMPLOYEES'
for index, supplier in df.iterrows():
    employees = str(supplier['EMPLOYEES RANGE']).split('-')
    
    if len(employees) > 1:
        df.at[index, 'MIN_EMPLOYEES'], df.at[index, 'MAX_EMPLOYEES'] = employees
        
    elif supplier['EMPLOYEES RANGE'] == '>5000':
        df.at[index, 'MIN_EMPLOYEES'] = '5000'

df.drop('EMPLOYEES RANGE', axis=1, inplace=True)

df.to_csv('BenettonSuppliersCleaned5.0.csv', index=False)

df.head()
    
    
```
